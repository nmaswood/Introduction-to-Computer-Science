# 3. Binary : To be on or to not be on

## 3.1 A simple example

__How many chicks do you count in the following picture__

<img src="assets/chicks.png" alt="Drawing" style="width: 400px;"/>

Most (all?) people would say :

__5__


However if you were a computer you would say:

___101___

Both answers are right


__This is because you can represent the same number in different number systems.__

For example take the number ___2048___

<img src="assets/2048.jpg" alt="Drawing" style="width: 800px;height:400px"/>


In base ten this number is represented as __2048__ because


\begin{align}
2048 &= 2 \times  1000 + 0 \times 100 + 4 \times 10 + 8 \times 1\\
2048 &= 2 \times  10^3 + 0 \times 10^2 + 4 \times 10^1 + 8 \times 10^0\\
\end{align}

In base two this number is represented as __100000000000__ because

\begin{align}
2048 &= 1 \times 2^{11} + 0 \times 2^{10}  + \ldots + 0 \times  2^{0}
\end{align}

## 3.2 Definition of Binary

Use of the base-2 number system is referred to as __binary__. Binary uses two mutually exclusive states to represent information. A binary number is made up of elements called bits where each bit can be in one of the two possible states. Generally, we represent them with the numerals 1 and 0. We also talk about them being true and false. Electrically, the two states might be represented by high and low voltages or some form of switch turned on or off.

## 3.3 Uses of Binary


### Math

Computers store information in binary and perform arithmetic on numbers in binary form. For example look how one can add two numbers together in binary.


<img src="assets/add.png" alt="Drawing" style="width: 800px;height:400px"/>

### Letters

Binary can also be used to represent, letters and characters. An encoding system called ASCII is used to map between simple characters and numbers represented in binary.

<img src="assets/ascii.png" alt="Drawing" style="width: 800px;height:400px"/>

### Colors

Numbers in the form of binary can also be used to represent colors. The image on your computer is made from millions of tiny blocks called pixels (picture elements). Each one of these is effectively a separate red, blue, or green light that can be switched on or off very rapidly to make the moving color picture. 

Numbers ranging from 1 to 255 control the intensity of each light and thus different number combinations referring the intensity of the RGB lights can be used to make images.
