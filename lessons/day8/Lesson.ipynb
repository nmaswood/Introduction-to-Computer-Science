__What is the problem with this method of storing data?__

* There is a lot of repetition
* How to uniquely identify transaction?

__How to fix this?__

* Store data in something called a relation/table.
    * Each entity with have it own relation
    * User Relation
    * Merchant Relation
    * Transaction Relation
    
a __relation__ is a data structure which consists of a heading and an unordered set of tuples
* Give each item a unique numerical id

The kind of data organization we just constructed is called a relational database

<img src="assets/db.png" alt="" style="width: 800px;" /> 

Wherein data is organized into one or more tables (or "relations") of columns and rows, with a unique key identifying each row. Rows are also called records or tuples.

While there are several models for organizing data relational model is most popular. SQL which you may have heard of is language designed for interacting with data stored in relation models.

<h3 align = "center"> Misc issues </h3>

* What happens if deposits money and your computer shutdowns before the money is deposited?
* What happens if a user and their partner withdraw money from the exact same time?
    * E.g. two customers withdrawing funds from the same account at the same time - account has \$500 in it, and they withdraw $100 and $50. The result could be \$350, \$400 or \$450 if no protection.
* Can you keep their data safe?
    * Hackers
    * Back ups
    * Every user of the system should be able to access only the data they are permitted to see.
        * E.g. payroll people only handle employee records, and cannot see customer accounts; tellers only access account data and cannot see payroll data.
* Concurrency
* Data in different files.
* Data in different formats.
* Data may be required to satisfy constraints.

Databases through various methods, help with all of these problems and are reasons for their existence.
